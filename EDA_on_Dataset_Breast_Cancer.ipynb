```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Breast Cancer Wisconsin.csv')
df
```

```python
# 1. Basic Info
print("🔹 First 5 Rows:")
print(df.head())
print("\n🔹 Dataset Info:")
print(df.info())
print("\n🔹 Shape:", df.shape)
```

```python
# 2. Missing Values
print("\n🔹 Missing Values:")
print(df.isnull().sum())
```

```python
# 3. Data Types
print("\n🔹 Data Types:")
print(df.dtypes)
```

```python
# 4. Statistical Summary
print("\n🔹 Statistical Summary:")
print(df.describe())
```

```python
# 5. Target Column Distribution
print("\n🔹 Target Class Distribution:")
if 'diagnosis' in df.columns:
    print(df['diagnosis'].value_counts())
    sns.countplot(x='diagnosis', data=df)
    plt.title("Diagnosis Count")
    plt.show()
```

```python
# 6. Histograms
df.hist(bins=20, figsize=(20, 15), color='skyblue')
plt.suptitle("Histograms of Features")
plt.show()
```

```python
# 7. Boxplots to detect outliers
plt.figure(figsize=(15, 8))
df.drop('id', axis=1, errors='ignore').boxplot(rot=90)
plt.title("Boxplots of Features")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 8. Correlation Matrix
plt.figure(figsize=(12,10))
# Exclude non-numeric columns like 'diagnosis' before calculating correlation
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 9. Pairplot (on a few selected features if many)
selected_cols = df.columns[2:6].tolist() + ['diagnosis'] if 'diagnosis' in df.columns else df.columns[:5]
sns.pairplot(df[selected_cols], hue='diagnosis' if 'diagnosis' in df.columns else None)
plt.suptitle("Pairplot of Selected Features")
plt.show()
```
